```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
output_folder = "output"
all_data = []

def read_data(output_folder):
    for filename in os.listdir(output_folder):
        if filename.endswith(".txt"):
            filepath = os.path.join(output_folder, filename)
            with open(filepath, 'r') as f:
                lines = f.readlines()
                # Extract relevant data from the file
                num_nurses = int(lines[0].split(": ")[1])
                num_patients = int(lines[1].split(": ")[1])
                num_iterations = int(lines[2].split(": ")[1])
                num_particles = int(lines[3].split(": ")[1])
                solution_type = lines[4].strip()
                distances = eval(lines[6].strip())  # Convert string representation of list to list
                patients_list = eval(lines[8].strip())  # Convert string representation of list to list
                overall_fitness_score = float(lines[10].split(": ")[1])
                qualification_score = float(lines[11].split(": ")[1])
                timetable_compliance_score = float(lines[12].split(": ")[1])
                total_distance = float(lines[13].split(": ")[1])

                # Store data in a dictionary
                data = {
                    "filename": filename,
                    "num_nurses": num_nurses,
                    "num_patients": num_patients,
                    "num_iterations": num_iterations,
                    "num_particles": num_particles,
                    "solution_type": solution_type,
                    "distances": distances,
                    "patients_list": patients_list,
                    "overall_fitness_score": overall_fitness_score,
                    "qualification_score": qualification_score,
                    "timetable_compliance_score": timetable_compliance_score,
                    "total_distance": total_distance
                }

                # Append data to the list
                all_data.append(data)
    return all_data

all_data = read_data(output_folder)
filename = [data["filename"] for data in all_data]
num_nurses = [data["num_nurses"] for data in all_data]
num_patients = [data["num_patients"] for data in all_data]
num_iterations = [data["num_iterations"] for data in all_data]
num_particles = [data["num_particles"] for data in all_data]
distances = [data["distances"] for data in all_data]
patients_list = [data["patients_list"] for data in all_data]
overall_fitness_score = [data["overall_fitness_score"] for data in all_data]
qualification_score = [data["qualification_score"] for data in all_data]
timetable_compliance_score = [data["timetable_compliance_score"] for data in all_data]
total_distance = [data["total_distance"] for data in all_data]

def plot_data_for_patients(num_patients_filter):
    filtered_data = [data for data in all_data if data["num_patients"] == num_patients_filter]
    if not filtered_data:
        print(f"No data available for {num_patients_filter} patients.")
        return

    num_particles_filtered = [data["num_particles"] for data in filtered_data]
    overall_fitness_score_filtered = [data["overall_fitness_score"] for data in filtered_data]
    sorted_data = sorted(zip(num_particles_filtered, overall_fitness_score_filtered))
    num_particles_sorted, overall_fitness_score_sorted = zip(*sorted_data)

    # Plotting
    plt.figure(figsize=(8, 6))
    plt.plot(num_particles_sorted, overall_fitness_score_sorted, marker='o', linestyle='-')
    plt.title(f'Evolution of num_particles vs overall_fitness_score for {num_patients_filter} patients')
    plt.xlabel('num_particles')
    plt.ylabel('overall_fitness_score')
    plt.grid(True)
    plt.show()
    
for num_patients in [30, 35, 40, 100]:
    plot_data_for_patients(num_patients)

```
